# Missing data
Part of data exploration is finding where there are problems in the data. This includes NaNs, Infs, and other missing values. No matter how you handle missing values, it is important to understand why those values are missing. The most straightforward way to handle missing data is to simply remove those data points. However, missing values may not be distributed at random, and could themselves be predictive. Because it is difficult to know whether the missing values are informative, dropping data points is rarely a good choice.

Another simple strategy is to replace the missing values with the mean, however if there are many outliers this may give poor results. A better option is to replace the values with the median. Even better is to use a more sophisticated method, such as regressing the missing values.

There is a good write up about missing values on [Wikipedia](https://en.wikipedia.org/wiki/Missing_data), and a Python implementation of several imputation algorithms can be found [here](https://github.com/hammerlab/fancyimpute).

# Class balance

It can be a problem when a dataset has imbalanced classes, where one outcome is significantly more common in the data. In the worst case scenario, the learning algorithm can simply predict the most common class with probability 1 and do an acceptable job. Making sure the classes are balanced in the training set will avoid this.

One way to balance the classes is to sample from the underrepresented class with replacement. Having many identical data points isn't ideal, but is an improvement over imbalanced classes. A more sophisticated approach is to create synthetic data points from the existing data points. A popular algorithm to do this is `SMOTE`, a short description can be found [here](https://www.cs.cmu.edu/afs/cs/project/jair/pub/volume16/chawla02a-html/node6.html).

```python
df_titanic['Survived'].value_counts()
```

In this case our classes are roughly balanced. 

If the classes were not balanced, oversampling with replacement can be done with `RandomOverSampler` which is available in the `imblearn` package. `SMOTE` is also available in the `imblearn` package.

# Standardization

Many machine learning algorithms require the data to be **isotropic**, or to have roughly equal variance.

Using `describe()` we see that the variance is different for our numerical values, such as Age and Fare (as you might expect). **Standardization** is the process of transforming a variable to have zero mean and unit variance, and is done by subtracting the mean and dividing by the standard deviation. **Normalization** is sometimes used to mean the same thing, but more accurately is scaling a feature to lie between 0 and 1. Normalizing a variable is done by subtracting the minimum value and dividing by the difference between the maximum and minimum value.

```python
numeric_columns = ['Age','Fare','Parch', 'Pclass','SibSp']
df_titanic_transformed[numeric_columns] = df_titanic_transformed[numeric_columns].apply(lambda x: (x - np.mean(x))/(np.std(x)) , axis=0)
```

# Categorical Features

The best way to know which featuers are categorical is to look at the original data source, although the number of unique values can give a sense as well.

```python
unique_values = df_titanic.apply(lambda x: x.nunique(),axis=0)
sns.barplot(df_titanic.columns,unique_values)
```

In this case the only categorical values are 'Embarked' and 'Sex'.

Categorical features need to be handled carefully, if you simply number them numerically, you are implictly making an assumption about the distance between categories. For example, if we had a feature encoding a subjects favorite fruit, and the options were ('apple','orange','banana'), we could try encoding these numerically as (0,1,2). However, this implicitly tells an ML algorithm that the "distance" between apples and oranges is the same as between oranges and bananas, but apples and bananas are twice as far apart. Obviously, that doesn't make sense any way you slice it.

Instead, we use one-hot encoding to represent each category as a feature. This creates a new column for each category where there is a 1 if that category was present and a 0 if not. `sklearn` has a one-hot encoding function, but we can also do all the processing in `pandas` using `get_dummies`:

```python
df_titanic_transformed = pd.get_dummies(df_titanic_transformed,columns=['Sex','Embarked'])
```

# Feature reduction

Although feature engineering can significantly improve performance, too many features will lead to overfitting of the data. In the worst case scenario, where there are as many features as there are data points, an ML algorithm can memorize the training set, and have terrible generalization to a test set.

One way to check for redundant features is to compute the correlation between features. Although including highly correlated features in a model doesn't bias the results, it does make interpretation of some models difficult. In this case the highest Pearson's correlation is about 0.55, which is low enough to not worry too much. # Test data set

You remembered to keep a separate test dataset right? Luckily in this case the data came pre-split and we've only been working with the training set. **It is incredibly important to split out a test set that is statistically representative of the training set and keep that entirely separate from all model training.** 

The testing set should **only** be checked after all training is complete. Using the training set to help choose parameters while training is one type of *data leakage* and is a serious problem. It will make your predictions appear abnormally good, and can lead to serious problems if your company takes action on this anomalous prediction.

The next step would be testing on the test set. Although the predictors are available, Kaggle makes you submit to check against the targets. However, we would need to transform the test data the same way we transformed the training data, so below is a compilation of the operations we did above. 

```python
def process_titanic_data(df):
    df['Age'].fillna(df['Age'].median(),inplace=True)
    df['Title'] = df['Name'].apply(lambda x: x.split(',')[1].split()[0])
    numeric_columns = ['Age','Fare','Parch', 'Pclass','SibSp']
    df[numeric_columns] = df[numeric_columns].apply(lambda x: (x - np.mean(x))/(np.std(x)) , axis=0)
    df['Cabin_exists'] = df['Cabin'].apply(lambda x: 0 if str(x) == 'nan' else 1)
    df = pd.get_dummies(df,columns=['Sex','Embarked', 'Title'])
    return df
```
